```python
import requests
import numpy as np
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt
```

```python
url = 'http://localhost:8080/change_background'
```

```python
with open('test_im.jpeg', 'rb') as f:
    image_data = f.read()
```

```python
response = requests.post(url, files={'image': image_data})
```

```python
new_image_data = BytesIO(response.content)
new_image = Image.open(new_image_data)
changed_background = np.array(new_image)
```

```python
plt.imshow(changed_background)
```
